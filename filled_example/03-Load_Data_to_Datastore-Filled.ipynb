# Load your Data with Azure Storage Explorer

When working with new data in Azure Machine Learning Service, the most reliable data to load data into your Storage Account is with Azure Storage Explorer.  It's simple to use, familiar interface that lets you upload files to your Blob Storage Account or Azure Data Lake Gen 2 in the same way that you would upload files to a directory on your local machine.

To Learn more and Download Azure Storage Explorer, click here: https://azure.microsoft.com/en-us/features/storage-explorer/
    

Navigate to your Subscription using the left hand panel.<br>
Once you have selected your subcription, select your Storage Account.<br>
Once you have selected your Storage Account, select the appropriate Container.  You can also create a new Container by right clicking "Blob Containers".<br>
Once you have selected a Container, navigate to the appropriate Directory on your Storage Account.  You can create a new Directory by clicking "New Folder."<br>
Once you have selected a Directory on your Container, click the Upload button.<br>
Navigate to the file on you wish to upload from your local machine and select it to upload your data to your Storage Account.
